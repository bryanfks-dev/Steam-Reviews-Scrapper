```python
# Import required modules
# pandas is a module to work with dataframes
import pandas as pd
```

```python
# csv_file_path is a variable that contains the path of the CSV file that the program will create
# to store the result of scraped data
#
# Replace the value of the variable with the path where you want to store the CSV file
csv_file_path: str = "iphone_15_pro_reviews.csv"
```

```python
# Read the CSV file and store the data in a dataframe
df = pd.read_csv(csv_file_path)

# Print the csv file as a dataframe
df
```

```python
# Display the dataframe informations
df.info()
```

```python
# pre_df is a dataframe that contains the data after applying the preprocessing steps
pre_df = pd.get_dummies(df, columns=["Stars"], drop_first=True)

# Display the preprocessed dataframe
pre_df.head()
```

```python
# sklean.model_selection is a module that contains the train_test_split function
# train_test_split is a function that splits the data into training and testing sets
from sklearn.model_selection import train_test_split
```

```python
X = pre_df.drop("not.fully.paid", axis=1)
y = pre_df["not.fully.paid"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```
